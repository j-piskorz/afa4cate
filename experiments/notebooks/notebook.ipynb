```python
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from PIL import Image
from pathlib import Path

# Step 1: Load the MNIST dataset
# Using torchvision to download and load the dataset
mnist_data = datasets.MNIST(root="/home/azureuser/localfiles/afa4cate/afa4cate/datasets/mnist", train=True, download=True)
image, label = mnist_data[5]  # Load the first image and label

# Display the original image
plt.imshow(image, cmap='gray')
plt.title(f'Original MNIST Image (28x28) - Label: {label}')
plt.axis('off')
plt.show()
```

```python
# Step 2: Define a function to resize images to 8x8
def resize_image(image, size=(8, 8)):
    """
    Resizes a given PIL image to the specified size.
    
    Parameters:
    - image: PIL Image, the input image to resize.
    - size: tuple, the desired output size (width, height).
    
    Returns:
    - resized_image: PIL Image, the resized image.s
    """
    resized_image = image.resize(size, Image.LANCZOS)
    return resized_image

# Step 3: Resize the MNIST image to 8x8 and display the resized version
resized_image = resize_image(image)

# Convert to display using matplotlib
plt.imshow(resized_image, cmap='gray')
plt.title(f'Resized MNIST Image (8x8) - Label: {label}')
plt.axis('off')
plt.show()
```

```python
import numpy as np

# Step 1: Create a mask for a single 8x8 image
single_mask = np.zeros((8, 8), dtype=int)

# Set the border pixels to 1
single_mask[0, :] = 1          # Top row
single_mask[-1, :] = 1         # Bottom row
single_mask[:, 0] = 1          # Left column
single_mask[:, -1] = 1         # Right column

# Step 2: Flatten the mask to a 1D array of length 64
flattened_mask = single_mask.flatten()  # Shape: (64,)

# Step 3: Repeat the mask for all images in the dataset
n_samples = 10000  # Number of images in your dataset
full_mask = np.tile(flattened_mask, (n_samples, 1))  # Shape: (10000, 64)

# Now, 'full_mask' is a (10000, 64) array with the desired masking
```
